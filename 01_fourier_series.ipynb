```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap

plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams.update({'font.size': 18})

dx = 0.001
L = np.pi
x = L * np.arange(-1 + dx, 1 + dx, dx)
n = len(x)
nquart = int(np.floor(n / 4))

# Define hat function
f = np.zeros_like(x)
f[nquart:2 * nquart] = (4 / n) * np.arange(1, nquart + 1)
f[2 * nquart:3 * nquart] = np.ones(nquart) - (4 / n) * np.arange(0, nquart)

# Function to compute inner product using dot product
def inner_product(f1, f2):
    return np.dot(f1, f2) * dx


fig, ax = plt.subplots()
ax.plot(x, f, '-', color='k', label="Original function")


name = "Accent"
cmap = get_cmap("tab10")
colors = cmap.colors
ax.set_prop_cycle(color=colors)

A0 = inner_product(f, np.ones_like(x))
fFS = A0 / 2  

A = np.zeros(20)
B = np.zeros(20)

for k in range(1, 21):
    cos_k = np.cos(k * np.pi * x / L)
    sin_k = np.sin(k * np.pi * x / L)

    A[k-1] = inner_product(f, cos_k)
    B[k-1] = inner_product(f, sin_k)

    fFS += A[k-1] * cos_k + B[k-1] * sin_k
    ax.plot(x, fFS, '-', label=f'k={k}')

ax.legend()
plt.title("Fourier Series Approximation of Hat Function")
plt.show()
```
